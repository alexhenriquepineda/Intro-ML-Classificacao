# Introdução a Modelos de Classificação - Parte 4

```python
import pandas as pd
import seaborn as sns

from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
dados = pd.read_csv('https://gist.githubusercontent.com/guilhermesilveira/1b7d5475863c15f484ac495bd70975cf/raw/16aff7a0aee67e7c100a2a48b676a2d2d142f646/projects.csv')
```

```python
sns.scatterplot(x='expected_hours', y='price', hue='unfinished', data=dados)
```

# Rescaling

```python
x = dados[['expected_hours', 'price']]
y = dados['unfinished']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, stratify=y)

scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)

model = LinearSVC()
model.fit(x_train, y_train)
previsoes = model.predict(x_test)
acuracia = accuracy_score(y_test, previsoes) * 100
acuracia
```

```python
import numpy as np
previsoes_de_base = np.ones(540)
acuracia = accuracy_score(y_test, previsoes_de_base) * 100
acuracia
```
